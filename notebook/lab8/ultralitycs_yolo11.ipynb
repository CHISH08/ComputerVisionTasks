## –ó–∞–≥—Ä—É–∂–∞–µ–º –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–µ –≤–µ—Å–∞ –º–æ–¥–µ–ª–∏ yolov11n

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
```

```python
# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
results = model.train(
    data='/kaggle/input/diordataset/Diorkaggle.yaml',
    epochs=8,
    imgsz=640,
    batch=32,
    lr0=3e-3,
    name="yolov11_dior",
    freeze=22, # –∑–∞–º–æ—Ä–∞–∂–∏–≤–∞–µ–º 22 —Å–ª–æ—è –∏–∑ 29
    save_period=4,
    val=True,

    # –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏–∏:
    hsv_h=0.015,   # –ò–∑–º–µ–Ω–µ–Ω–∏–µ –æ—Ç—Ç–µ–Ω–∫–∞ (hue)
    hsv_s=0.7,     # –ò–∑–º–µ–Ω–µ–Ω–∏–µ –Ω–∞—Å—ã—â–µ–Ω–Ω–æ—Å—Ç–∏ (saturation)
    hsv_v=0.4,     # –ò–∑–º–µ–Ω–µ–Ω–∏–µ —è—Ä–∫–æ—Å—Ç–∏ (value)
    flipud=0.0,    # –û—Ç–∫–ª—é—á–µ–Ω–∏–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–≥–æ –æ—Ç—Ä–∞–∂–µ–Ω–∏—è
    fliplr=0.5,    # –ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–µ –æ—Ç—Ä–∞–∂–µ–Ω–∏–µ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 50%
    mosaic=1.0,    # –í–∫–ª—é—á–µ–Ω–∏–µ –º–æ–∑–∞–∏–∫–∏ (–æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ 4 –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π)
    degrees=0.0,   # –ü–æ–≤–æ—Ä–æ—Ç –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π (–≤ –≥—Ä–∞–¥—É—Å–∞—Ö)
    translate=0.1, # –°–¥–≤–∏–≥ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –Ω–∞ 10%
)
```

```python
import os
```

```python
image_dir = "/kaggle/input/diordataset/Dior/Dior/images/val"
image_files = [os.path.join(image_dir, f) for f in os.listdir(image_dir)[:5]]

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Å —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
results = model.predict(
    source=image_files,  # –°–ø–∏—Å–æ–∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π [[4]]
    save=True,           # –°–æ—Ö—Ä–∞–Ω–∏—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å –±–æ–∫—Å–∞–º–∏ [[3]]
    save_dir="output",   # –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è [[5]]
    conf=0.4            # –ü–æ—Ä–æ–≥ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏
)
```
